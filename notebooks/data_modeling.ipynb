# Data Modeling

## ONE Data

```python
# Reeding the data
capitas_pagadas = pd.read_excel('../data/raw/one/afiliados-número-cápitas-cápitas-pagadas-sfs-según-régimen-2005-2023.xlsx', engine='openpyxl')

capitas_pagadas
```

```python
# Filtering the unwanted rows.
capitas_pagadas = capitas_pagadas.iloc[2:-3]
```

```python
raw_column_name = ''

# Iterating over the rows.
for i in range(capitas_pagadas.shape[0]):
    # If cell corresponds to a title is saved.
    if capitas_pagadas.iloc[i, 0] is not np.nan:
        raw_column_name = capitas_pagadas.iloc[i, 0]
    
    # Creating the column name.
    column_name = f'{raw_column_name} ({capitas_pagadas.iloc[i, 1]})'

    # If column name exist add an index number to differentiate it from the other columns.
    if column_name in capitas_pagadas.iloc[:, 1].values: 
        column_name += f' {i}'

    # Title is added to the sub title to have the column name.
    capitas_pagadas.iloc[i, 1] = column_name

# Transposing and filtering the data.
capitas_pagadas = capitas_pagadas.iloc[:, 1:].T.reset_index(drop=True)
capitas_pagadas
```

```python
# Obtaining the column headers.
headers = capitas_pagadas.iloc[0]
# Replacing the headers.
capitas_pagadas.columns = headers
# Dropping unwanted rows.
capitas_pagadas.drop(labels=0, inplace=True)

capitas_pagadas
```

```python
# Filling the missing years.
capitas_pagadas.iloc[:, 0] = capitas_pagadas.iloc[:, 0].fillna(method='ffill')
```

```python
# Filtering rows.
capitas_pagadas = capitas_pagadas[(capitas_pagadas.iloc[:, 1].str.strip() == 'Cuarto trimestre') | (capitas_pagadas.iloc[:, 1].isna())]
capitas_pagadas
```

```python
# Dropping unwanted columns
capitas_pagadas.drop('Años (nan) 1', axis=1, inplace=True)
capitas_pagadas
```

## CNSS

```python
data = pd.read_excel('../data/raw/cnss/Excel-–-Regimen-Contributivo-Recaudos-y-Pagos-Seguro-Familiar-de-Salud-Seguros-Julio-2003-a-Noviembre-2024.xlsx')
data.head()
```

```python
data2 = pd.read_excel('../data/raw/cnss/Excel-–-Regimen-Subsidiado-Aportes-Pagos-Seguro-Familiar-de-Salud-Enero-2004-octubre-2024-a-Noviembre-2024-1.xlsx', engine='openpyxl')
data2.head()
```

```python
# Standardizing columns names.
columns = ['mes' if column.lower().strip() == 'meses' else column.lower().strip() for column in data.columns]
# Updating the columns names.
data.columns = columns
# Creating a new column combining the year and month.
data['fecha'] = data['año'].astype('str') + ' ' + data['mes']

# Dropping the month and year
data.drop(['año', 'mes'], axis=1, inplace=True)
data.head()
```

```python
# Standardizing columns names.
columns = ['mes' if column.lower().strip() == 'meses' else column.lower().strip() for column in data2.columns]
# Updating the columns names.
data2.columns = columns
# Creating a new column combining the year and month.
data2['fecha'] = data2['año'].astype('str') + ' ' + data2['mes']
# Dropping the month and year
data2 = data2.iloc[:, 2:]
data2.head()
```

```python
# Merging datasets.
data3 = pd.merge(data, data2, on='fecha')

# Splitting the date into year and month.
data3[['anio', 'mes']] = data3['fecha'].str.split(' ', expand=True, n=1)

# Dropping "date" columns.
data3.drop('fecha', axis=1, inplace=True)
data3.head()
```
